### Задача 5 Распределение ресурсов: ДП

```python
import numpy as np
from scipy.interpolate import interp1d

Z, Np, a5, b5 = 12, 3, 0.55, 0.45
gx = np.array([0, 3, 6, 9, 12], dtype=float)

g = lambda Y: 3 * Y + 0.25 * Y ** 2
h = lambda z: 4 * z + 0.2 * z ** 2

F5, Y5 = {1: np.zeros_like(gx)}, {1: np.zeros_like(gx)}
for i, X in enumerate(gx):
    Ys = np.arange(0, X + 1e-9, 1.0)
    v = g(Ys) + h(X - Ys)
    j = int(v.argmax())
    F5[1][i], Y5[1][i] = v[j], Ys[j]

for k in range(2, Np + 1):
    F5[k], Y5[k] = np.zeros_like(gx), np.zeros_like(gx)
    fp = interp1d(gx, F5[k-1], kind='linear', fill_value='extrapolate')
    for i, X in enumerate(gx):
        Ys = np.arange(0, X + 1e-9, 1.0)
        v = g(Ys) + h(X - Ys) + fp(a5 * Ys + b5 * (X - Ys))
        j = int(v.argmax())
        F5[k][i], Y5[k][i] = v[j], Ys[j]

Xc, inc = float(Z), 0.0

for k in range(Np, 0, -1):
    y = float(interp1d(gx, Y5[k])(Xc))
    y = max(0, min(y, Xc))
    
    inc += g(y) + h(Xc - y)
    Xc = a5 * y + b5 * (Xc - y)

print(f'Суммарный доход = {round(inc, 3)}')
```
