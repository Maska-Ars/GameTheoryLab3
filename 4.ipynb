### Задача 4 Замена оборудования: Ротация микросервисов

```python
Tmax, Nseasons, t0 = 5, 4, 0

R = lambda t: (35 - 3.5 * t) if 0 <= t <= Tmax else 0.0
C = lambda t: 5 + 2.5 * t 

ages = range(0, Tmax + Nseasons + 2)
Fk, pol = {1: {}}, {1: {}}

for t in ages:
    keep, repl = R(t), R(0) - C(t)
    Fk[1][t], pol[1][t] = (keep, 'K') if keep >= repl else (repl, 'R')

for k in range(2, Nseasons + 1):
    Fk[k], pol[k] = {}, {}
    for t in ages:
        keep = R(t) + Fk[k - 1].get(t + 1, 0.0)
        repl = R(0) - C(t) + Fk[k - 1][1]
        Fk[k][t], pol[k][t] = (keep, 'K') if keep >= repl else (repl, 'R')
print(f'F4(t0) = {Fk[Nseasons][t0]}')
```
