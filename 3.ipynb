### Задача 3 Задача Джонсона: Конвейер CI/CD

```python
from collections import deque

A = [6, 2, 5, 9, 3, 7, 4]
B = [5, 7, 2, 1, 6, 3, 8]
n = len(A)
 
def johnson(A, B):
    head, tail, remaining = deque(), deque(), set(range(n))
    while remaining:
        i_best, v_best, stage = None, None, None
        
        for i in remaining:
            if v_best is None or A[i] < v_best: 
                v_best, i_best, stage = A[i], i, 'A'
            if B[i] < v_best:
                v_best, i_best, stage = B[i], i, 'B'
        
        if stage == 'A':
            head.append(i_best)
        else:
            tail.appendleft(i_best)
        
        remaining.discard(i_best)

    return list(head) + list(tail)
 
def metrics(order):
    tA = 0
    tB = 0
    
    for i in order:
        tA += A[i]; tB = max(tA, tB) + B[i]
    
    return tB, tB - sum(B)
 
opt = johnson(A, B)

print(f'Оптимальный порядок: {[i+1 for i in opt]}')
print(f'     Общее время, простой B')
print(f'Исходный:    {metrics(range(n))}')
print(f'Оптимальный: {metrics(opt)}')
 
```
